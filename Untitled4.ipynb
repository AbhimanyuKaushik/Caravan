```python
import pandas as pd, numpy as np, math, matplotlib.pyplot as plt
import warnings, statsmodels.api as sm
warnings.filterwarnings("ignore")

from sklearn.preprocessing import MinMaxScaler, StandardScaler, LabelEncoder
from sklearn.metrics import mean_squared_error

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout, Bidirectional, Input
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
```

```python
df = pd.read_csv("FinalBossMerged(SIH).csv")
df["Date"] = pd.to_datetime(df["Date"])
df = df.sort_values(["DISTRICT", "Date"])

le_district = LabelEncoder()
df["DISTRICT_ENC"] = le_district.fit_transform(df["DISTRICT"])

df["Month"] = df["Date"].dt.month

# ---- target normalisation (0-1) ----
scaler_wl = MinMaxScaler(feature_range=(0, 1))
df["WL_norm"] = scaler_wl.fit_transform(df[["WL(mbgl)"]])

# ---- coordinates standardisation ----
scaler_geo = StandardScaler()
df[["LAT_norm", "LON_norm"]] = scaler_geo.fit_transform(df[["LATITUDE", "LONGITUDE"]])

# --------------------------------------------------------------
# 3. Sequence creator (longer history = better performance)
# --------------------------------------------------------------
def create_sequences_multi(df, time_step=60):
    X, y = [], []
    for district in df["DISTRICT_ENC"].unique():
        temp = df[df["DISTRICT_ENC"] == district].reset_index(drop=True)
        vals = temp[["WL_norm","LAT_norm","LON_norm","DISTRICT_ENC","Month"]].values
        for i in range(len(temp) - time_step - 1):
            X.append(vals[i:(i+time_step), :])
            y.append(vals[i+time_step, 0])          # WL_norm
    return np.array(X), np.array(y)

time_step = 60
X, y = create_sequences_multi(df, time_step)

train_size = int(len(X) * 0.8)
X_train, X_test = X[:train_size], X[train_size:]
y_train, y_test = y[:train_size], y[train_size:]

# --------------------------------------------------------------
# 4. Stronger LSTM (Bidirectional + deeper) + Input layer
# --------------------------------------------------------------
model = Sequential([
    Input(shape=(time_step, X.shape[2])),               # <-- fixes warning
    Bidirectional(LSTM(64, return_sequences=True)),
    Dropout(0.2),
    Bidirectional(LSTM(32, return_sequences=False)),
    Dropout(0.2),
    Dense(32, activation='relu'),
    Dense(1)                                            # predicts WL_norm
])

model.compile(optimizer='adam', loss='mse')

early_stop = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)
reduce_lr  = ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=2, min_lr=1e-6)

model.fit(X_train, y_train,
          epochs=50,
          batch_size=64,
          validation_data=(X_test, y_test),
          callbacks=[early_stop, reduce_lr],
          verbose=1)

# --------------------------------------------------------------
# 5. SARIMA on the *raw* series (last district)
# --------------------------------------------------------------
last_district = df["DISTRICT_ENC"].iloc[-1]
temp = df[df["DISTRICT_ENC"] == last_district].copy()
temp = temp.groupby("Date", as_index=False)["WL(mbgl)"].mean()
temp = temp.set_index("Date").asfreq("MS")
ts_raw = temp["WL(mbgl)"].interpolate()

sarima = sm.tsa.statespace.SARIMAX(ts_raw,
                                   order=(1,1,1),
                                   seasonal_order=(1,1,1,12),
                                   enforce_stationarity=False,
                                   enforce_invertibility=False)
sarima_res = sarima.fit(disp=False)

split_idx = int(len(ts_raw) * 0.8)
sarima_pred_raw = sarima_res.predict(start=split_idx, end=len(ts_raw)-1).values

# --------------------------------------------------------------
# 6. Bring everything to **normalized** space
# --------------------------------------------------------------
lstm_test_pred_norm = model.predict(X_test, verbose=0).flatten()   # already 0-1
sarima_pred_norm    = scaler_wl.transform(sarima_pred_raw.reshape(-1,1)).flatten()
y_test_norm         = y_test[-len(sarima_pred_norm):]               # ground-truth (norm)

# --------------------------------------------------------------
# 7. Ensemble weighting (grid search on *normalized* RMSE)
# --------------------------------------------------------------
weights = np.linspace(0, 1, 101)
best_rmse, best_w = float('inf'), None

for w_lstm in weights:
    w_sarima = 1 - w_lstm
    blend = w_lstm * lstm_test_pred_norm[-len(sarima_pred_norm):] + w_sarima * sarima_pred_norm
    rmse = math.sqrt(mean_squared_error(y_test_norm, blend))
    if rmse < best_rmse:
        best_rmse, best_w = rmse, (w_lstm, w_sarima)

print("\n=== NORMALISED PERFORMANCE ===")
print(f"LSTM-only  RMSE (norm): {math.sqrt(mean_squared_error(y_test_norm, lstm_test_pred_norm[-len(sarima_pred_norm):])):.4f}")
print(f"SARIMA-only RMSE (norm): {math.sqrt(mean_squared_error(y_test_norm, sarima_pred_norm)):.4f}")
print(f"Best weights (LSTM, SARIMA): {best_w}")
print(f"Best ensemble RMSE (norm): {best_rmse:.4f}")

# --------------------------------------------------------------
# 8. Future forecast (13 months) – stay normalised until plot
# --------------------------------------------------------------
future_steps = 13
last_seq = X[-1:].copy()
future_lstm_norm = []

for i in range(future_steps):
    nxt = model.predict(last_seq, verbose=0)[0,0]
    future_lstm_norm.append(nxt)

    # next row (WL stays normalised)
    new_row = np.array([nxt,
                        last_seq[0,-1,1],   # LAT_norm
                        last_seq[0,-1,2],   # LON_norm
                        last_seq[0,-1,3],   # DISTRICT_ENC
                        (df["Month"].iloc[-1] + i + 1) % 12 or 12])
    last_seq = np.append(last_seq[:,1:,:], new_row.reshape(1,1,-1), axis=1)

future_lstm_norm = np.array(future_lstm_norm)

# SARIMA future → normalised
sarima_future_raw = sarima_res.forecast(steps=future_steps).values
sarima_future_norm = scaler_wl.transform(sarima_future_raw.reshape(-1,1)).flatten()

# Blend
future_blend_norm = best_w[0] * future_lstm_norm + best_w[1] * sarima_future_norm

# ---- ONLY NOW inverse-transform for the plot ----
future_blend_mbgl = scaler_wl.inverse_transform(future_blend_norm.reshape(-1,1)).flatten()
future_dates = pd.date_range(start="2025-09-01", periods=future_steps, freq="MS")

# --------------------------------------------------------------
# 9. Plot (mbgl) – show normalised RMSE in title
# --------------------------------------------------------------
plt.figure(figsize=(12,6))
plt.plot(future_dates, future_blend_mbgl,
         marker='o', color='red', linewidth=2.5,
         label=f"Ensemble Forecast (norm RMSE = {best_rmse:.3f})")
plt.title("Groundwater Level Prediction – Sep 2025 to Sep 2026")
plt.xlabel("Date"); plt.ylabel("Water Level (mbgl)")
plt.grid(True); plt.legend()
plt.tight_layout()
plt.show()
```
